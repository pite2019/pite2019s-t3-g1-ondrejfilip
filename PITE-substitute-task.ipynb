```python
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score
from sklearn import datasets, linear_model
import sklearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
boston = load_boston()
df = pd.DataFrame(data=boston.data)
```

```python
df.head()
```

```python
df.hist(column=[5,12], bins=30)
```

```python
X = df.loc[:, [5,12]]
y = boston.target
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
predict = regr.predict(X)

plt.scatter(X[5], y)
plt.title('Scatter plot of the column "5" and target')
plt.show()

plt.plot(X[5], predict, linewidth=1)
plt.title('Plot of the column "5" and predicted values of the model')
plt.show()
```

```python
predict = regr.predict(X)

plt.scatter(X[12], y)
plt.title('Scatter plot of the column "12" and target')
plt.show()

plt.plot(X[12], predict, linewidth=1)
plt.title('Plot of the column "12" and predicted values of the model')
plt.show()
```

# Results

```python
regr.coef_
```

The model has two coeficients (the closer to the zero the better).

```python
r2_score(y, predict)
```
